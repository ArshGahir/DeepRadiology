```python
# split data into train/test sets
import pandas as pd
from sklearn.model_selection import train_test_split

# load data
df = pd.read_csv('train.csv')

# add column to identify split/train
df['is_train'] = True

# split data with stratification
train_df, test_df = train_test_split(df, test_size=0.2, random_state=8, stratify=df['class_id'])

# Update the 'is_train' column for the test set
test_df['is_train'] = False

# Concatenate the train and test sets back into a single DataFrame
final_df = pd.concat([train_df, test_df])

print('Data split and labeling complete')

final_df.head()
```

```python
import os
import shutil
import pandas as pd

def move_images(df, source_folder, train_folder, test_folder):
    for _, row in df.iterrows():
        image_id = row['image_id']
        is_train = row['is_train']

        source_path = os.path.join(source_folder, f"{image_id}.tiff")
        destination_folder = train_folder if is_train else test_folder

        # move image
        shutil.move(source_path, os.path.join(destination_folder, f"{image_id}.tiff"))

if __name__ == "__main__":
    # source and destination folders
    source_folder = 'C:/Users/Alex/Desktop/SENECA/Semester 9/BDC800_Capstone/Capstone Project/DeepRadiology/dataset/images/train_og'
    train_folder = 'C:/Users/Alex/Desktop/SENECA/Semester 9/BDC800_Capstone/Capstone Project/DeepRadiology/dataset/images/train'
    test_folder = 'C:/Users/Alex/Desktop/SENECA/Semester 9/BDC800_Capstone/Capstone Project/DeepRadiology/dataset/images/test'
    
    # move images to proper locations
    move_images(final_df, source_folder, train_folder, test_folder)
    print('Images moved to appropriate folders')
```
